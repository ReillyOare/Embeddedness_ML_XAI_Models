# Renamed EPA Dataset Merging

```python
# Read in libs
import pandas as pd
```

## Step 1. Read in Data

```python
# Read in 2008-2009 datasets
epa_0809_siteinfo = pd.read_excel('0809phabmed_renamed.xlsx')
epa_0809_phab = pd.read_excel('0809phabmed_renamed.xlsx')
epa_0809_land = pd.read_excel('epa_0809_land_renamed.xlsx')
epa_0809_chem = pd.read_excel('epa_0809_chem_renamed.xlsx')
# Read in 2013-2013 datasets
epa_1314_siteinfo = pd.read_excel('nrsa1314_siteinfo_renamed.xlsx')
epa_1314_phab = pd.read_excel('nrsa1314_phab_renamed.xlsx')
epa_1314_land = pd.read_excel('epa_1314_land_renamed.xlsx')
epa_1314_chem = pd.read_excel('epa_1314_chem_renamed.xlsx')
# Read in 2018-2019 datasets
epa_1819_siteinfo = pd.read_excel('nrsa_1819_siteinfo_renamed.xlsx')
epa_1819_phab = pd.read_excel('nrsa_1819_phab_renamed.xlsx')
epa_1819_land = pd.read_excel('epa_1819_land_renamed.xlsx')
epa_1819_chem = pd.read_excel('epa_1819_chem_renamed.xlsx')
```

## Step 2. Drop rows without the target

```python
# Select for only datapoints with target value, Mean streambed embeddedness (%)
epa_0809_emd = epa_0809_phab.dropna(subset = ['Mean streambed embeddedness (%)'], axis = 0)
epa_1314_emd = epa_1314_phab.dropna(subset = ['Mean streambed embeddedness (%)'], axis = 0)
epa_1819_emd = epa_1819_phab.dropna(subset = ['Mean streambed embeddedness (%)'], axis = 0)
```

```python
# Print number of rows in each dataset
print("Shape of epa 0809:", epa_0809_emd.shape)
print("Shape of epa 1314:", epa_1314_emd.shape)
print("Shape of epa 1819:", epa_1819_emd.shape)
```

# Step 3. Merge invidual years

### 2008-2009

```python
#Merge 2008 - 2009 datasets
#Find first set of common columns
common0809_1 = list(set(epa_0809_emd.columns).intersection(epa_0809_land.columns))
#Merge 2008 - 2009 embeddedness with land data
epa_0809_mrg1 = epa_0809_emd.merge(epa_0809_land,
                                  on = common0809_1,
                                  how = 'left')
#Print first 2008 - 2009 dataframe
epa_0809_mrg1
```

```python
#Find third set of common columns
common0809_2 = list(set(epa_0809_mrg1.columns).intersection(epa_0809_chem.columns))
#Merge 2008 - 2009 embeddedness with chem data
epa_0809_mrg2 = epa_0809_mrg1.merge(epa_0809_chem,
                                  on = common0809_2,
                                  how = 'left')
#Print second 2008-2009 dataframe
epa_0809_mrg2
```

### 2013-2014

```python
#Merge 2013 - 2014 datasets
#Find first set of common columns
common1314_1 = list(set(epa_1314_emd.columns).intersection(epa_1314_land.columns))
#Merge 2013 - 2014 embeddedness with land data
epa_1314_mrg1 = epa_1314_emd.merge(epa_1314_land,
                                  on = common1314_1,
                                  how = 'left')
#Print first 2013 - 2014 dataframe
epa_1314_mrg1
```

```python
#Find second set of common columns
common1314_2 = list(set(epa_1314_mrg1.columns).intersection(epa_1314_chem.columns))
#Merge 2013 - 2014 embeddedness with chem data
epa_1314_mrg2 = epa_1314_mrg1.merge(epa_1314_chem,
                                  on = common1314_2,
                                  how = 'left')
#Print first 2013 - 2014 dataframe
epa_1314_mrg2
```

### 2018-2019

```python
#Find first set of common columns
common1819_1 = list(set(epa_1819_emd.columns).intersection(epa_1819_siteinfo.columns))
#Merge 2018 - 2019 embeddedness with site info data
epa_1819_mrg1 = epa_1819_emd.merge(epa_1819_siteinfo,
                                  on = common1819_1,
                                  how = 'left')
#Print second 2018 - 2019 dataframe
epa_1819_mrg1
```

```python
# Find second set of common columns
common1819_2 = list(set(epa_1819_mrg1.columns).intersection(epa_1819_land.columns))
#Merge 2018 - 2019 embeddedness with land data
epa_1819_mrg2 = epa_1819_mrg1.merge(epa_1819_land,
                                  on = common1819_2,
                                  how = 'left')
#Print second 2018 - 2019 dataframe
epa_1819_mrg2
```

```python
# Find third set of common columns
common1819_3 = list(set(epa_1819_mrg2.columns).intersection(epa_1819_chem.columns))
#Merge 2018 - 2019 embeddedness with chem data
epa_1819_mrg3 = epa_1819_mrg2.merge(epa_1819_chem,
                                  on = common1819_3,
                                  how = 'left')
#Print second 2018 - 2019 dataframe
epa_1819_mrg3
```

## Step 3. Merge Model Building datasets

### Step 3a. Drop columns with more than 5% of data missing from each year's datasets

```python
pd.set_option('display.max_rows', 5)
```

```python
missing_0809 = epa_0809_mrg2.isnull().mean()
missing_1314 = epa_1314_mrg2.isnull().mean()
missing_1819 = epa_1819_mrg3.isnull().mean()
```

```python
# Drop columns where the missing percentages are more than 10% of the column (0.05
epa_0809_cln = epa_0809_mrg2.drop(columns = missing_0809[missing_0809 > 0.05].index)
epa_1314_cln = epa_1314_mrg2.drop(columns = missing_1314[missing_1314 > 0.05].index)
epa_1819_cln = epa_1819_mrg3.drop(columns = missing_1819[missing_1819 > 0.05].index)
```
